# Creazione del corpus

```python
import pandas as pd
```

```python
df_corpus = pd.read_csv("trainset/desegma-it.subTaskA.shared.train.0923-1220.csv")
```

```python
df_corpus
```

```python
df_corpus["label"].value_counts()
```

```python
df_corpus[df_corpus["label"] == 0]
```

### Creazione Training Set dal train Set (1000 e 1000)

```python
train_class0 = df_corpus[df_corpus["label"] == 0].sample(n=1000, random_state = 42)
```

```python
train_class1 =  df_corpus[df_corpus["label"] == 1].sample(n=1000, random_state = 42)
```

```python
df_train = pd.concat([train_class0, train_class1])
df_train
```

```python
df_train["label"].value_counts()
```

### Creazione del validation set dal training set: 1000 documenti (500 e 500)

```python
val_class0 = df_corpus[df_corpus["label"] == 0].drop(train_class0.index).sample(n=500, random_state=42)
```

```python
val_class1 = df_corpus[df_corpus["label"] == 1].drop(train_class1.index).sample(n=500, random_state=42)
```

```python
df_val = pd.concat([val_class0, val_class1])
df_val
```

```python
df_val["label"].value_counts()
```

### Creazione del Test set dal Test set

```python
df_test_completo = pd.read_csv("testset/desegma-it.subTaskA.with_labels.test.1117_1835.csv")
```

```python
df_test_completo
```

```python
df_test_completo["label"].value_counts()
```

```python
test_class0 = df_test_completo[df_test_completo["label"] == 0].sample(n=500, random_state = 42)
test_class1 = df_test_completo[df_test_completo["label"] == 1].sample(n=500, random_state = 42)
```

```python
df_test = pd.concat([test_class0, test_class1])
df_test["label"].value_counts()
```

```python
df_train.to_csv("df_train.csv")
df_val.to_csv("df_val.csv")
df_test.to_csv("df_test.csv")
```
